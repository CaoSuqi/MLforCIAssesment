```python
# ============================================================
# 1) 配置类 (CFG)
# ============================================================
class CFG:
    # 路径配置
    NPZ_PATH = "F:/PyWorkspace/suqi/AutoEncoder/Results_route1/segment_raw_features.npz"
    LABELS_CSV = "F:/PyWorkspace/suqi/AutoEncoder/data/labels.csv"
    #REPORT_PDF  = "F:/PyWorkspace/suqi/AutoEncoder/Results_xgb_baseline/zh_Modeling_Mild_Server_Report.pdf"
    #METRICS_CSV  = "F:/PyWorkspace/suqi/AutoEncoder/Results_xgb_baseline/zh_Metrics_Mild_Server_Summary.csv"
    EPS = 1e-8
```

```python
# --------------------------
# 1) Baseline (median-of-mice) for healthy day0
# --------------------------
def _iqr(x, axis=0):
    return np.quantile(x, 0.75, axis=axis) - np.quantile(x, 0.25, axis=axis)

def fit_population_healthy_baseline_median_of_mice(
    X_raw: np.ndarray,
    labels_df: pd.DataFrame,
    train_idx: np.ndarray,
    by_electrode: bool = True,
):
    """
    用训练集中的健康(day==0)估计群体基线，策略=median-of-mice：
      - 先按 mouse 对健康样本取 median 向量
      - 再对这些“鼠代表向量”取 median/IQR

    by_electrode=True: 每个electrode单独估计一套 (median, IQR)
    """
    idx = np.asarray(train_idx)

    # 健康定义：day==0（与你实验一致）
    healthy_mask = (labels_df.loc[idx, "day"].values == 0)
    healthy_idx = idx[healthy_mask]
    if len(healthy_idx) == 0:
        raise ValueError("No healthy(day==0) samples in training split.")

    mice = labels_df.loc[healthy_idx, "mouse"].astype(str).values
    elec = labels_df.loc[healthy_idx, "electrode"].astype(int).values
    Xh = X_raw[healthy_idx]  # (Nh, F)

    F = X_raw.shape[1]

    if by_electrode:
        E = int(labels_df["electrode"].max()) + 1
        base_med = np.zeros((E, F), dtype=np.float32)
        base_iqr = np.zeros((E, F), dtype=np.float32)

        for e in np.unique(elec):
            # 在该电极上，按鼠聚合
            mlist = np.unique(mice[elec == e])
            M = []
            for m in mlist:
                Xm = Xh[(elec == e) & (mice == m)]
                M.append(np.median(Xm, axis=0))
            M = np.stack(M, axis=0)  # (#mice, F)

            base_med[e] = np.median(M, axis=0)
            base_iqr[e] = _iqr(M, axis=0) + EPS

        return {"median": base_med, "iqr": base_iqr, "by_electrode": True}

    else:
        mlist = np.unique(mice)
        M = []
        for m in mlist:
            Xm = Xh[mice == m]
            M.append(np.median(Xm, axis=0))
        M = np.stack(M, axis=0)

        base_med = np.median(M, axis=0).astype(np.float32)
        base_iqr = (_iqr(M, axis=0) + EPS).astype(np.float32)
        return {"median": base_med, "iqr": base_iqr, "by_electrode": False}

def transform_with_baseline(
    X_raw: np.ndarray,
    labels_part: pd.DataFrame,
    baseline: dict,
):
    """X_raw -> Z-healthy"""
    if baseline["by_electrode"]:
        med = baseline["median"]
        iqr = baseline["iqr"]
        elec = labels_part["electrode"].astype(int).values
        Z = (X_raw - med[elec]) / (iqr[elec] + EPS)
    else:
        Z = (X_raw - baseline["median"]) / (baseline["iqr"] + EPS)
    return Z.astype(np.float32)
```

```python
def eval_binary_mild_vs_sev(
    X_raw, labels_df, feature_names,exclude_mice='B1',seed=0
):
    """
    只保留 label in {1,2}. 重新编码：MILD=0, SEV=1
    每次留一只鼠做测试（严格不使用测试鼠day0拟合baseline）。
    去除B1老鼠
    """
    keep_mask = (labels_df["mouse"].astype(str) != exclude_mice).values
    X_f = X_raw[keep_mask]
    labels_f = labels_df[keep_mask].reset_index(drop=True)
    
    mice_sorted = sorted(labels_f["mouse"].astype(str).unique().tolist())
    print("Mouse list (sorted):", mice_sorted)

    rows = []
    cm_sum = np.zeros((2, 2), dtype=int)

    for k, test_mouse in enumerate(mice_sorted):
        test_mask_full = (labels_f["mouse"].astype(str).values == test_mouse)
        train_mask_full = ~test_mask_full

        # 训练/测试样本：仅取 MILD/SEV
        tr_idx = np.where(train_mask_full & labels_f["label"].isin([1, 2]).values)[0]
        te_idx = np.where(test_mask_full & labels_f["label"].isin([1, 2]).values)[0]

        # y 映射：MILD=0, SEV=1
        ytr = (labels_f.loc[tr_idx, "label"].values == 2).astype(int)
        yte = (labels_f.loc[te_idx, "label"].values == 2).astype(int)

        # baseline 拟合：只用训练鼠的 day0（健康）
        baseline_train_idx = np.where(train_mask_full)[0]  # 用训练鼠全部索引去筛 day0
        baseline = fit_population_healthy_baseline_median_of_mice(
            X_f, labels_f, train_idx=baseline_train_idx, by_electrode=True
        )
        Xtr = transform_with_baseline(X_f[tr_idx], labels_f.iloc[tr_idx].reset_index(drop=True), baseline)
        Xte = transform_with_baseline(X_f[te_idx], labels_f.iloc[te_idx].reset_index(drop=True), baseline)

        model = make_xgb_binary(seed=seed)
        model.fit(Xtr, ytr)

        proba = model.predict_proba(Xte)[:, 1]
        pred = (proba >= 0.5).astype(int) #0.5

        bal_acc = balanced_accuracy_score(yte, pred)
        rec = recall_score(yte, pred, average=None, labels=[0, 1])  # [MILD, SEV]
        try:
            auc = roc_auc_score(yte, proba)
        except Exception:
            auc = np.nan

        cm = confusion_matrix(yte, pred, labels=[0, 1])
        cm_sum += cm

        rows.append({
            "test_mouse": test_mouse,
            "balanced_acc": bal_acc,
            "recall_MILD": rec[0],
            "recall_SEV": rec[1],
            "auc": auc,
            "n_test": len(te_idx),
        })

        print(f"test_mouse={test_mouse} | "
              f"balAcc={bal_acc:.3f} recall(M,S)=[{rec[0]:.3f},{rec[1]:.3f}] AUC={auc:.3f} n_test={len(te_idx)}")

    df = pd.DataFrame(rows)
    print("\nBinary Summary mean±std across mice:")
    for col in ["balanced_acc", "recall_MILD", "recall_SEV", "auc"]:
        print(f"{col:>12s}: {df[col].mean():.3f} ± {df[col].std(ddof=1):.3f}")

    print("\nBinary Confusion Matrix SUM [rows=true, cols=pred] labels=[MILD(0), SEV(1)]:")
    print(cm_sum)

    return df, cm_sum
```

```python
# --------------------------
# 3) XGBoost model config
# --------------------------
# grid search
def make_xgb_binary(seed=0):
    return XGBClassifier(
        objective="binary:logistic",
        n_estimators=100,#150,200,600,
        learning_rate=0.2,#0.15,0.1,0.05
        max_depth=3,
        subsample=0.8,#0.7,0.9
        colsample_bytree=0.9, #0.8 0.7 0.6
        reg_lambda=1.0, #2 3
        min_child_weight=1.0, #2 3
        gamma=0, #0.1 0.2
        random_state=seed,
        n_jobs=-1,
        eval_metric="logloss",
    )
```
