```python
import os
import json
from pathlib import Path
from dataclasses import dataclass
import numpy as np
import pandas as pd
from scipy.signal import butter, filtfilt, welch
from scipy.stats import skew, kurtosis
from tqdm.auto import tqdm
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
@dataclass
class CFG:
    DATA_DIR: str = r"F:/PyWorkspace/suqi/AutoEncoder/data/data_flat"
    LABELS_CSV: str = r"F:/PyWorkspace/suqi/AutoEncoder/data/labels.csv"
    OUT_DIR: str = r"F:/PyWorkspace/suqi/AutoEncoder/MLforCIAssesment/Results"

    fs: int = 500
    filter_low: float = 1.0
    filter_high: float = 100.0
    butter_order: int = 4

    win_sec: float = 4.0
    hop_sec: float = 2.0

    bands = {
        "band_1_10": (1, 10),
        "band_10_20": (10, 20),
        "band_20_50": (20, 50),
        "band_50_100": (50, 100),
        "band_1_50": (1, 50),
    }

    # label index
    label_to_name = {0: "NO", 1: "MILD", 2: "SEVERE"}
```

```python
# -----------------------------
# 1) IO helpers
# -----------------------------
def ensure_dir(p: str):
    Path(p).mkdir(parents=True, exist_ok=True)

def read_signal_xlsx(
    xlsx_path: str,
    channel: int | None = None,
    return_time: bool = False,
):
    df = pd.read_excel(xlsx_path, engine="openpyxl")

    # time
    time_col = None
    if "Time_s" in df.columns:
        time_col = "Time_s"

    # signal列
    sig_col = None

    if channel is None:
        # 优先找 Channel_ 开头列
        ch_cols = [c for c in df.columns if isinstance(c, str) and c.startswith("Channel_")]
        if len(ch_cols) == 1:
            sig_col = ch_cols[0]
        elif len(ch_cols) >= 2:
            # 多个Channel列的情况：默认取第二列（更符合你描述的两列文件）
            sig_col = df.columns[1]
        else:
            # fallback：取第二列
            if df.shape[1] < 2:
                raise ValueError(f"Excel列数不足2列：{xlsx_path}")
            sig_col = df.columns[1]
    else:
        cand1 = f"Channel_{channel:02d}"
        cand2 = f"Channel_{channel}"
        if cand1 in df.columns:
            sig_col = cand1
        elif cand2 in df.columns:
            sig_col = cand2
        else:
            raise KeyError(f"找不到 {cand1} 或 {cand2} 于文件：{xlsx_path}. 实际列：{list(df.columns)}")

    x = pd.to_numeric(df[sig_col], errors="coerce").to_numpy(dtype=float)
    x = np.nan_to_num(x, nan=0.0, posinf=0.0, neginf=0.0)

    if return_time:
        if time_col is None:
            raise KeyError(f"未找到 Time_s 列：{xlsx_path}")
        t = pd.to_numeric(df[time_col], errors="coerce").to_numpy(dtype=float)
        t = np.nan_to_num(t, nan=np.nan)
        return t, x

    return x

def bandpass_filter(x: np.ndarray, fs: int, low: float, high: float, order: int = 4) -> np.ndarray:
    nyq = 0.5 * fs
    low_n = low / nyq
    high_n = high / nyq
    b, a = butter(order, [low_n, high_n], btype="bandpass")
    return filtfilt(b, a, x)
```

```python
# -----------------------------
# 2) Feature extraction
# -----------------------------
def zero_crossing_rate(x: np.ndarray) -> float:
    s = np.sign(x)
    return float(np.mean(s[1:] * s[:-1] < 0))

def hjorth_params(x: np.ndarray):
    # Activity, Mobility, Complexity
    x = np.asarray(x)
    dx = np.diff(x)
    ddx = np.diff(dx)
    var0 = np.var(x) + 1e-12
    var1 = np.var(dx) + 1e-12
    var2 = np.var(ddx) + 1e-12
    activity = var0
    mobility = np.sqrt(var1 / var0)
    complexity = np.sqrt(var2 / var1) / (mobility + 1e-12)
    return float(activity), float(mobility), float(complexity)

def bandpower(psd_f, psd_pxx, fmin, fmax):
    mask = (psd_f >= fmin) & (psd_f <= fmax)
    if mask.sum() < 2:
        return 0.0
    return float(np.trapezoid(psd_pxx[mask], psd_f[mask]))

def spectral_entropy(psd_pxx):
    p = psd_pxx / (np.sum(psd_pxx) + 1e-12)
    p = np.clip(p, 1e-12, 1.0)
    return float(-np.sum(p * np.log(p)))

def extract_features_one_window(xw: np.ndarray, fs: int, bands: dict):
    # time-domain
    mean_abs = float(np.mean(np.abs(xw)))
    rms = float(np.sqrt(np.mean(xw ** 2) + 1e-12))
    std = float(np.std(xw))
    p2p = float(np.max(xw) - np.min(xw))
    sk = float(skew(xw))
    ku = float(kurtosis(xw))
    zcr = zero_crossing_rate(xw)
    act, mob, comp = hjorth_params(xw)

    # freq-domain via Welch
    f, pxx = welch(xw, fs=fs, nperseg=min(len(xw), 1024))
    mask = (f >= 1) & (f <= 100)
    f2, pxx2 = f[mask], pxx[mask]
    total_1_100 = float(np.trapezoid(pxx2, f2)) if len(f2) > 1 else 0.0

    feats = {
        "mean_abs": mean_abs,
        "rms": rms,
        "std": std,
        "p2p": p2p,
        "skew": sk,
        "kurtosis": ku,
        "zcr": zcr,
        "hjorth_activity": act,
        "hjorth_mobility": mob,
        "hjorth_complexity": comp,
        "spec_entropy_1_100": spectral_entropy(pxx2) if len(pxx2) > 0 else 0.0,
        "total_power_1_100": total_1_100,
    }

    # band powers + relative
    for k, (b0, b1) in bands.items():
        bp = bandpower(f2, pxx2, b0, b1)
        feats[f"power_{k}"] = bp
        feats[f"rel_power_{k}"] = bp / (total_1_100 + 1e-12)

    feats["log_energy"] = float(np.log1p(np.sum(xw ** 2)))

    return feats

def extract_features_per_file(x: np.ndarray, fs: int, win_sec: float, hop_sec: float, bands: dict):
    win = int(win_sec * fs)
    hop = int(hop_sec * fs)
    if len(x) < win:
        # 太短就直接算全段
        feats_list = [extract_features_one_window(x, fs, bands)]
    else:
        feats_list = []
        for s in range(0, len(x) - win + 1, hop):
            xw = x[s:s + win]
            feats_list.append(extract_features_one_window(xw, fs, bands))

    df = pd.DataFrame(feats_list)

    # 聚合：median + IQR（鲁棒）
    med = df.median(axis=0)
    iqr = (df.quantile(0.75) - df.quantile(0.25))

    out = {}
    for col in df.columns:
        out[f"{col}_med"] = float(med[col])
        out[f"{col}_iqr"] = float(iqr[col])
    return out

```

```python
# -----------------------------
# 3) Build dataset: segment-level
# -----------------------------
from tqdm import tqdm
def build_segment_level_npz(cfg):
    """
    生成并保存：
      segment_raw_features.npz   (X_raw)
    要求 labels.csv 至少包含列：
      file, mouse, day, label, electrode, rep
    """

    out_dir = Path(cfg.OUT_DIR)
    out_dir.mkdir(parents=True, exist_ok=True)

    labels_df = pd.read_csv(cfg.LABELS_CSV)

    # ---------- mouse / label index ----------
    label_to_name = getattr(cfg, "label_to_name", {0: "NO", 1: "MILD", 2: "SEVERE"})
    name_to_label = {v: k for k, v in label_to_name.items()}

    mice = sorted(labels_df["mouse"].astype(str).unique().tolist())
    mouse_to_idx = {m: i for i, m in enumerate(mice)}
    idx_to_mouse = {i: m for m, i in mouse_to_idx.items()}

    (out_dir / "label_index.json").write_text(
        json.dumps({"label_to_name": label_to_name, "name_to_label": name_to_label}, ensure_ascii=False, indent=2),
        encoding="utf-8"
    )
    (out_dir / "mouse_index.json").write_text(
        json.dumps({"mouse_to_idx": mouse_to_idx, "idx_to_mouse": idx_to_mouse}, ensure_ascii=False, indent=2),
        encoding="utf-8"
    )

    # ---------- extract RAW features ----------
    feature_names = None
    X_raw_list = []

    N = len(labels_df)

    # 用结构化数组存 meta（避免 object/pickle）
    meta_dtype = np.dtype([
        ("file", "U260"),
        ("mouse", "U40"),
        ("day", "i4"),
        ("label", "i4"),
        ("electrode", "i4"),
        ("rep", "i4"),
    ])
    meta = np.zeros(N, dtype=meta_dtype)

    y = np.zeros(N, dtype=np.int64)
    groups_mouse = np.zeros(N, dtype=np.int64)

    for i, row in tqdm(labels_df.iterrows(), total=len(labels_df), desc="Extracting features"):
        rel_file = str(row["file"])
        xlsx_path = os.path.join(cfg.DATA_DIR, rel_file)
        if not os.path.exists(xlsx_path):
            raise FileNotFoundError(f"Missing file: {xlsx_path}")

        # 读信号（两列：Time_s + Channel_XX）
        x = read_signal_xlsx(xlsx_path, return_time=False)

        # 滤波
        x = bandpass_filter(x, cfg.fs, cfg.filter_low, cfg.filter_high, cfg.butter_order)

        # 提特征：滑窗 -> 聚合(median + IQR)
        feats = extract_features_per_file(
            x=x,
            fs=cfg.fs,
            win_sec=cfg.win_sec,
            hop_sec=cfg.hop_sec,
            bands=cfg.bands
        )

        if feature_names is None:
            feature_names = list(feats.keys())
        else:
            # 防止某次返回 dict key 顺序/集合不同
            if set(feats.keys()) != set(feature_names):
                missing = set(feature_names) - set(feats.keys())
                extra = set(feats.keys()) - set(feature_names)
                raise ValueError(f"Feature keys mismatch at row {i}. missing={missing}, extra={extra}")

        X_raw_list.append([feats[k] for k in feature_names])

        # meta / labels
        meta[i]["file"] = rel_file
        meta[i]["mouse"] = str(row["mouse"])
        meta[i]["day"] = int(row["day"])
        meta[i]["label"] = int(row["label"])
        meta[i]["electrode"] = int(row["electrode"])
        meta[i]["rep"] = int(row["rep"])

        y[i] = int(row["label"])
        groups_mouse[i] = mouse_to_idx[str(row["mouse"])]

    X_raw = np.asarray(X_raw_list, dtype=np.float32)
    feature_names_arr = np.array(feature_names, dtype="U120")  

    # 保存 RAW
    raw_npz_path = out_dir / "segment_raw_features.npz"
    np.savez(
        raw_npz_path,
        X=X_raw,
        y=y,
        groups_mouse=groups_mouse,
        meta=meta,                    
        feature_names=feature_names_arr
    )
    print(f"[OK] Saved RAW : {raw_npz_path}  X={X_raw.shape}")
    print(f"[OK] N={N}, F={X_raw.shape[1]}, mice={mice}")

    return str(raw_npz_path)
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

if __name__ == "__main__":
    raw_path = build_segment_level_npz(CFG)
```

### outlier mouse detect

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.covariance import EllipticEnvelope
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import scipy.stats as stats

def detect_outlier_mice(X_raw, labels_df, feature_names=None, contamination=0.2,seed=0):
    """
    按小鼠进行离群检测（每个小鼠用中位数特征表示）
    
    参数:
        X_raw: 原始特征矩阵 (n_samples, n_features)
        labels_df: 包含mouse, day, electrode, label等信息的DataFrame
        feature_names: 特征名称列表
        contamination: 预计的离群比例
    """
    print("=" * 60)
    print("离群小鼠检测分析 (Robust Covariance - Elliptic Envelope)")
    print("=" * 60)
    
    # 1. 获取小鼠列表
    mice_list = labels_df['mouse'].unique()
    print(f"总小鼠数: {len(mice_list)}")
    print(f"小鼠列表: {sorted(mice_list)}")
    
    # 2. 按小鼠聚合特征（使用中位数）- 与原始代码一致
    mouse_features = []
    mouse_names = []
    mouse_sample_counts = []  # 每只鼠的样本数
    
    for mouse in sorted(mice_list):
        mask = (labels_df['mouse'] == mouse).values
        # 检查是否有足够的样本
        if mask.sum() < 3:
            print(f"警告: 小鼠 {mouse} 只有 {mask.sum()} 个样本，跳过")
            continue
        
        # 使用中位数聚合每只鼠的特征 - 关键步骤！
        mouse_feat = np.median(X_raw[mask], axis=0)
        mouse_features.append(mouse_feat)
        mouse_names.append(mouse)
        mouse_sample_counts.append(mask.sum())
    
    X_mouse = np.array(mouse_features)
    print(f"\n聚合后特征矩阵形状: {X_mouse.shape}")
    
    # 3. 运行Elliptic Envelope离群检测 
    try:
        elliptic = EllipticEnvelope(
            contamination=0.2,      # 固定为0.2，与你之前成功代码一致
            random_state=seed,        # 固定随机种子
            support_fraction=None   # 使用默认值
        )
        
        # 注意：这里使用原始数据，不进行标准化
        elliptic.fit(X_mouse)
        elliptic_labels = elliptic.predict(X_mouse)
        # 计算决策分数
        decision_scores = elliptic.decision_function(X_mouse)
        # 4. 输出结果 - 格式与之前一致
        print("\n" + "-" * 40)
        print("离群检测结果:")
        print("-" * 40)
        
        outlier_count = 0
        for i, (mouse, label, score, count) in enumerate(zip(mouse_names, elliptic_labels, decision_scores, mouse_sample_counts)):
            status = "⚠️ 离群" if label == -1 else "✓ 正常"
            print(f"  {mouse}: {status} (decision score = {score})")
            
            if label == -1:
                outlier_count += 1
        
        print("\n" + "-" * 40)
        print(f"总结: 检测到 {outlier_count} 只离群小鼠 (离群率: {outlier_count/len(mouse_names):.1%})")
        
        # 5. 返回结果
        outlier_indices = np.where(elliptic_labels == -1)[0]
        outlier_mice = [mouse_names[i] for i in outlier_indices]
        
        return {
            'mouse_names': mouse_names,
            'mouse_features': X_mouse,
            'mouse_sample_counts': mouse_sample_counts,
            'elliptic_labels': elliptic_labels,
            'decision_scores': decision_scores,
            'outlier_indices': outlier_indices,
            'outlier_mice': outlier_mice,
            'detector': elliptic,
            'feature_names': feature_names
        }
        
    except Exception as e:
        print(f"  检测失败: {e}")
        return None
```

```python
class CFG:
 NPZ_PATH = "F:/PyWorkspace/suqi/AutoEncoder/MLforCIAssesment/Results/segment_raw_features.npz"
 LABELS_CSV = "F:/PyWorkspace/suqi/AutoEncoder/MLforCIAssesment/data/labels.csv"
cfg=CFG()
d = np.load(cfg.NPZ_PATH, allow_pickle=False)
X_raw = d["X"].astype(np.float32)
y = d["y"].astype(np.int64)
groups_mouse = d["groups_mouse"].astype(np.int64)
labels = pd.read_csv(cfg.LABELS_CSV)
try:
    feature_names = [str(s) for s in d["feature_names"]]
except Exception:
    raise ValueError("NPZ文件中不包含feature_names，请重新保存")
```

```python
results = detect_outlier_mice(
    X_raw,
    labels_df=labels,
    feature_names=feature_names,  # 如果有特征名称可以提供
    contamination=0.2,    # 预计离群率
    seed=0
)
```
